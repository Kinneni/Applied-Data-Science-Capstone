```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the distance between two points on Earth using Haversine formula
    Returns distance in kilometers
    """
    # Approximate radius of Earth in km
    R = 6373.0
    
    # Convert coordinates to radians
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    
    # Calculate differences
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    distance = R * c
    return distance

# Example: Select KSC LC-39A launch site
launch_site_lat = 28.573255
launch_site_lon = -80.646895
launch_site_name = "KSC LC-39A"

print(f"Analyzing proximities for: {launch_site_name}")
print(f"Coordinates: ({launch_site_lat}, {launch_site_lon})")

# Create a map centered on the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Add a marker for the launch site (red marker)
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup=f"<b>{launch_site_name}</b>",
    icon=folium.Icon(color='red', icon='rocket', prefix='fa')
).add_to(site_map)

# Define proximities - example coordinates
# (In real scenario, you would have actual coordinates of nearby cities, railways, etc.)
proximities = [
    {'name': 'Orlando City', 'lat': 28.5383, 'lon': -81.3792, 'type': 'city'},
    {'name': 'Cape Canaveral', 'lat': 28.3922, 'lon': -80.6077, 'type': 'city'},
    {'name': 'Cocoa Beach', 'lat': 28.3200, 'lon': -80.6076, 'type': 'city'},
    {'name': 'Titusville', 'lat': 28.6122, 'lon': -80.8075, 'type': 'city'},
    {'name': 'Highway 1', 'lat': 28.5630, 'lon': -80.5770, 'type': 'highway'},
]

print("\n" + "="*70)
print("PROXIMITY ANALYSIS")
print("="*70)

# Calculate and display distances
for proximity in proximities:
    prox_name = proximity['name']
    prox_lat = proximity['lat']
    prox_lon = proximity['lon']
    prox_type = proximity['type']
    
    # Calculate distance
    distance = calculate_distance(launch_site_lat, launch_site_lon, prox_lat, prox_lon)
    
    print(f"{prox_name} ({prox_type}): {distance:.2f} km")
    
    # Choose marker color based on type
    if prox_type == 'city':
        marker_color = 'blue'
        marker_icon = 'building'
    elif prox_type == 'highway':
        marker_color = 'green'
        marker_icon = 'road'
    else:
        marker_color = 'gray'
        marker_icon = 'info-sign'
    
    # Add marker for proximity
    folium.Marker(
        [prox_lat, prox_lon],
        popup=f"<b>{prox_name}</b><br>Distance: {distance:.2f} km",
        icon=folium.Icon(color=marker_color, icon=marker_icon, prefix='fa')
    ).add_to(site_map)
    
    # Draw a line between launch site and proximity
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [prox_lat, prox_lon]],
        color='purple',
        weight=2,
        opacity=0.6,
        popup=f"Distance: {distance:.2f} km",
        dash_array='5, 5'
    ).add_to(site_map)

# Add distance circles (5km, 10km, 20km, 50km radii)
colors = ['green', 'orange', 'red', 'darkred']
radii = [5000, 10000, 20000, 50000]

for radius, color in zip(radii, colors):
    folium.Circle(
        [launch_site_lat, launch_site_lon],
        radius=radius,
        color=color,
        fill=False,
        weight=1,
        opacity=0.5,
        popup=f'{radius/1000} km radius'
    ).add_to(site_map)

print("="*70)
print("\n✓ Task 3 completed!")

# Display the map
site_map
```

```python
import folium
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# TASK 3: Calculate distances and create markers with lines
launch_site_lat = 28.573255
launch_site_lon = -80.646895
launch_site_name = "KSC LC-39A"

# Coastline coordinates
coastline_lat = 28.5630
coastline_lon = -80.5670

# Calculate distance
distance_to_coastline = calculate_distance(launch_site_lat, launch_site_lon, 
                                            coastline_lat, coastline_lon)

# Create map
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup=launch_site_name,
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(site_map)

# Create a marker with distance to coastline
folium.Marker(
    [coastline_lat, coastline_lon],
    popup=f"Coastline<br>Distance: {distance_to_coastline:.2f} km",
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Draw a line between the marker and the launch site
coordinates = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
lines = folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(lines)

# Save the map
site_map.save('task3_coastline_distance.html')
print(f"Distance to coastline: {distance_to_coastline:.2f} km")
print("✓ Map saved!")

site_map
```
